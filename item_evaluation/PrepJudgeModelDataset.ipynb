```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
ai_df = pd.read_csv("/home/aml7990/Code/creativity-item-generation/item_evaluation/Content Review/AllItemsToScore_rated.csv")
human_df = pd.read_csv("/home/aml7990/Code/creativity-item-generation/item_evaluation/Content Review/humanwrittenitemsrated.csv")
```

```python
# drop unneeded columns
ai_df = ai_df[["creative_scenario_round_4", "Complexity_rater5", "Difficulty_rater5","Accessibility_rater5","Controversial_rater5","Word List Inclusion_rater5","Leading_rater5",
            "Difficulty_rater4","Accessibility_rater4","Controversial_rater4","Word List Inclusion_rater4","Leading_rater4", "Complexity_rater4",
            "Difficulty_rater3","Accessibility_rater3","Controversial_rater3","Word List Inclusion_rater3","Leading_rater3", "Complexity_rater3",
            "Difficulty_rater2","Accessibility_rater2","Controversial_rater2","Word List Inclusion_rater2","Leading_rater2", "Complexity_rater2",
            "Difficulty_rater1","Accessibility_rater1","Controversial_rater1","Word List Inclusion_rater1","Leading_rater1", "Complexity_rater1",
            "Difficulty_raterGA1","Accessibility_raterGA1","Controversial_raterGA1","Word List Inclusion_raterGA1","Leading_raterGA1", "Complexity_raterGA1",
            "Difficulty_raterGA2","Accessibility_raterGA2","Controversial_raterGA2","Word List Inclusion_raterGA2","Leading_raterGA2", "Complexity_raterGA2",
               ]
               ]
```

```python
# clean df:
from math import floor
ai_df["complexity_aggregrate"] = -1
ai_df["difficulty_aggregrate"] = -1

for index, row in ai_df.iterrows():
    non_null_cols = row.dropna()
    # drop any non-acessible, leading, or controversial items (rated so by at least 1 ratrer)
    controverisal_ids = non_null_cols.index.str.contains("Controversial")
    if any(non_null_cols.loc[controverisal_ids]):
        # print("here")
        continue
    accessible_ids = non_null_cols.index.str.contains("Accessibility")
    if any(non_null_cols.loc[accessible_ids]):
        # print("here")
        continue
    leading_ids = non_null_cols.index.str.contains("Leading")
    if any(non_null_cols.loc[leading_ids]):
        # print("here")
        continue

    # compute the floor of the average complexity and difficulty
    complexity_ids = non_null_cols.index.str.contains("Complexity")
    complexity_agg = floor(non_null_cols.loc[complexity_ids].mean())
    ai_df.at[index, "complexity_aggregrate"] = complexity_agg
    
    difficulty_ids = non_null_cols.index.str.contains("Difficulty")
    difficulty_agg = floor(non_null_cols.loc[difficulty_ids].mean())
    ai_df.at[index, "difficulty_aggregrate"] = difficulty_agg
    # break
```

```python
ai_df = ai_df[(ai_df["complexity_aggregrate"] != -1) | (ai_df["difficulty_aggregrate"] != -1)]
```

```python
mask = ai_df["creative_scenario_round_4"].str.contains("Human: Word list:")
mask = mask.loc[mask == True]
```

```python
# drop items that don't follow correct format
ai_df = ai_df.drop(index=mask.index)
```

```python
# drop extra columns
ai_df = ai_df[["creative_scenario_round_4", "complexity_aggregrate", "difficulty_aggregrate"]]
```
