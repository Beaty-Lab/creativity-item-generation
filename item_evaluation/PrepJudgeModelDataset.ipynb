```python
# drop items that are leading or contain the word list
ai_df = ai_df.drop(index=ai_df[(ai_df["Leading_rater5"] == 1)].index)
ai_df = ai_df.drop(columns=["Leading_rater5", "Word List Inclusion_rater5"])
ai_df = ai_df.drop(index=ai_df[(ai_df["Leading_rater4"] == 1)].index)
ai_df = ai_df.drop(columns=["Leading_rater4", "Word List Inclusion_rater4"])
ai_df = ai_df.drop(index=ai_df[(ai_df["Leading_rater3"] == 1)].index)
ai_df = ai_df.drop(columns=["Leading_rater3", "Word List Inclusion_rater3"])
ai_df = ai_df.drop(index=ai_df[(ai_df["Leading_rater2"] == 1)].index)
ai_df = ai_df.drop(columns=["Leading_rater2", "Word List Inclusion_rater2"])
ai_df = ai_df.drop(index=ai_df[(ai_df["Leading_rater1"] == 1)].index)
ai_df = ai_df.drop(columns=["Leading_rater1", "Word List Inclusion_rater1"])
```

```python
ai_df
```

```python
# clean df:
from math import floor
ai_df["complexity_aggregrate"] = -1
ai_df["difficulty_aggregrate"] = -1

for index, row in ai_df.iterrows():
    non_null_cols = row.dropna()
    # drop any non-acessible or controversial items (rated so by at least 1 ratrer)
    controverisal_ids = non_null_cols.index.str.contains("Controversial")
    if any(non_null_cols.loc[controverisal_ids]):
        # print("here")
        continue
    accessible_ids = non_null_cols.index.str.contains("Accessibility")
    if any(non_null_cols.loc[accessible_ids]):
        # print("here")
        continue

    # compute the floor of the average complexity and difficulty
    complexity_ids = non_null_cols.index.str.contains("Complexity")
    complexity_agg = floor(non_null_cols.loc[complexity_ids].mean())
    ai_df.at[index, "complexity_aggregrate"] = complexity_agg
    
    difficulty_ids = non_null_cols.index.str.contains("Difficulty")
    difficulty_agg = floor(non_null_cols.loc[difficulty_ids].mean())
    ai_df.at[index, "difficulty_aggregrate"] = difficulty_agg
    # break
```

```python
ai_df = ai_df[(ai_df["complexity_aggregrate"] != -1) | (ai_df["difficulty_aggregrate"] != -1)]
```
