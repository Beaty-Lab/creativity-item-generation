```python
import seaborn as sns
import pandas as pd
from os.path import join
from matplotlib import pyplot as plt
```

```python
PATH = "C:/Users/aml7990/downloads"
no_balance_gpt4 = pd.read_csv(join(PATH,"No_balance_responses_quality_GPT-4o.csv"))
no_balance_gpt4["model"] = "GPT-4o"
no_balance_gemini = pd.read_csv(join(PATH,"No_balance_responses_quality_Gemini.csv"))
no_balance_gemini["model"] = "Gemini"
no_balance = pd.concat([no_balance_gpt4,no_balance_gemini])

half_gravity_gpt4 = pd.read_csv(join(PATH, "Half_gravity_responses_quality_GPT-4o.csv"))
half_gravity_gemini = pd.read_csv(join(PATH, "Half_gravity_responses_quality_Gemini.csv"))
half_gravity_gpt4["model"] = "GPT-4o"
half_gravity_gemini["model"] = "Gemini"
half_gravity = pd.concat([half_gravity_gpt4, half_gravity_gemini])
```

```python
sns.histplot(x=no_balance["Quality"],hue=no_balance["model"])
plt.title("No Balance")
```

```python
sns.histplot(x=half_gravity["Quality"],hue=half_gravity["model"])
plt.title("Half Gravity")
```
